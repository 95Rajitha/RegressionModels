```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataSet = pd.read_csv("Position_Salaries.csv")
dataSet.head(20)
```

```python
# we dont split data sets to test and trained sets since the we need to get the maximum training data for model trainning
x = dataSet.iloc[:,1:-1]
y = dataSet.iloc[:,-1]
print(y)
```

```python
from sklearn.linear_model import LinearRegression
linRegressor = LinearRegression()
linRegressor.fit(x,y)
```

```python
# this is the implementation to make the polynomial function accoradance with the multiple linear regression.
from sklearn.preprocessing import PolynomialFeatures
poliRegressor = PolynomialFeatures(degree=3)
x_poli = poliRegressor.fit_transform(x)
linRegressor2 = LinearRegression()
linRegressor2.fit(x_poli,y)
```

```python
# plotting linear regression 
plt.scatter(x,y,color='red')
plt.plot(x,linRegressor.predict(x),color='blue')
plt.title("truth checker")
plt.xlabel("position level")
plt.ylabel("salary")
plt.show
```

```python
#plpot results for the plinomial regression
plt.scatter(x,y,color="red")
plt.plot(x,linRegressor2.predict(x_poli),color="blue")
plt.title("truth checker with polinomial func")
plt.xlabel("position level")
plt.ylabel("salary")
plt.show()
```

```python
# Predicting new results using Linear regression
linRegressor.predict([[6.5]]) 
# this prediction is related to the simple linear regression and its accuracy is very low
```

```python
# prediction for the polinomial regression
# this aint working since we need to inpu the x^2 and so on depending upon the degree
# linRegressor2.predict([[6.5]])
linRegressor2.predict(poliRegressor.fit_transform([[6.5]]))
```
