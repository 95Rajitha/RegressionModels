```python
#Feature scaling doesn'nt needed here since even though it has different scales we use splitting techniques under both random forrest tree and decission tree algorithms 
#  for better accuracy we need more data in noth decission tree and Random forrest algos 

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataSet = pd.read_csv("Position_Salaries.csv")
x = dataSet.iloc[:,1:-1].values
y = dataSet.iloc[:,-1].values
```

```python
from sklearn import tree
# random state set to 0 for preventing the slightly doffernt results will happen when trainning the model
regressor = tree.DecisionTreeRegressor(random_state=0)
regressor.fit(x,y)
```

```python
regressor.predict([[6.5]])
```

```python
# for multi features this step should not set
x_grid = np.arange(min(x),max(x),0.1)
x_grid = x_grid.reshape((len(x_grid),1))
plt.scatter(x,y,color="red")
plt.plot(x_grid,regressor.predict(x_grid),color="blue")
plt.title("Decission Tree Reg for salaary predict")
plt.xlabel("level")
plt.ylabel("salary")
plt.show()


# note that
# when dealing with large data set  it will  be easy t 
```
